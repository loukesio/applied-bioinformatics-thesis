# Visualizations

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plot = sns.catplot(
    data=pd.DataFrame(
        columns=["Feature Selection Technique", "Machine Learning Model", "Accuracy"],
        data = [
            ["DE-PVALUE", "Logistic Regression", 0.96],
            ["DE-UNIVARIATE", "Logistic Regression", 0.96],
            ["PCA", "Logistic Regression", 0.83],

            ["DE-PVALUE", "Decision Tree", 0.90],
            ["DE-UNIVARIATE", "Decision Tree", 0.92],
            ["PCA", "Decision Tree", 0.70],

            ["DE-PVALUE", "Gradient Boosted Trees", 0.96],
            ["DE-UNIVARIATE", "Gradient Boosted Trees", 0.96],
            ["PCA", "Gradient Boosted Trees", 0.76],

            ["DE-PVALUE", "Random Forest", 0.96],
            ["DE-UNIVARIATE", "Random Forest", 0.94],
            ["PCA", "Random Forest", 0.83],

            ["DE-PVALUE", "Support Vector Machines", 0.96],
            ["DE-UNIVARIATE", "Support Vector Machines", 0.92],
            ["PCA", "Support Vector Machines", 0.83],
        ]
    ), 
    x="Feature Selection Technique", 
    y="Accuracy", 
    hue="Machine Learning Model", 
    kind="bar",
    palette=sns.color_palette(["#c9e3c2","#8bc8cb","#eccd80","#F5ab70","#9c87a1",])
)
plt.ylim(0.6, 1.0)
```
