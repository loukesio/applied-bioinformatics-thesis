```python
from adex.models import Condition, SequencingTechnique, TissueEnum, ConditionSequencingTissueDataLoader
from adex.mds import MdsHelper

from adex.models import TISSUE_COLORS, ConditionDataLoader

import logging
logging.basicConfig(format='%(levelname)s : %(message)s', level=logging.INFO)
logging.root.level = 20

files_path = "../data/adex-database/samples"
metadata_path = "../data/adex-database/metadata.csv"
datasets_info_path = "../data/adex-database/datasets_info.csv"
```

# Optimising the MDS 'n-components' parameter

```python
for condition in Condition:
    mds = MdsHelper(ConditionDataLoader(condition), files_path, metadata_path, datasets_info_path)
    mds.draw_components_stress_plot()
```

> For more convenient plotting 'n-components=2', 3 would be ideal.

# MDS per Condition 

```python
for condition in Condition:
    mds = MdsHelper(ConditionDataLoader(condition), files_path, metadata_path, datasets_info_path)

    mds.draw_2d(
        column_that_defines_colors="Condition",
        target_colors=[
            ("Healthy", "g"),
            (condition.name, "r")
        ]
    )
    
    mds.draw_2d(
        column_that_defines_colors="Tissue",
        target_colors=TISSUE_COLORS
    )
```

# MDS per Condition, Sequencing Technique, and Tissue

```python
for sequencing_technique in SequencingTechnique:
    for condition in Condition:
        for tissue in TissueEnum:
            mds = MdsHelper(
                ConditionSequencingTissueDataLoader(
                    condition=condition, 
                    sequencing_technique=sequencing_technique, 
                    tissue=tissue
                ), 
                files_path, 
                metadata_path, 
                datasets_info_path
            )

            mds.draw_2d(
                column_that_defines_colors="Condition",
                target_colors=[
                    ("Healthy", "g"),
                    (condition.name, "r")
                ]
            ) 
```
